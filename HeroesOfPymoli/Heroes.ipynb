```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head()
```

## Purchasing Analysis (Total)

```python
#total unique items
unique_items = purchase_data['Item ID'].nunique()

#average price
avg_price = (purchase_data['Price'].sum()/purchase_data['Price'].count()).round(2)

#total purchases made
total_purchases = purchase_data['Price'].count()

#total revenue
total_revenue = purchase_data["Price"].sum()

#build a summary data frame
total_analysis_df = pd.DataFrame({"Number of Unique Items": [unique_items], 
                              "Average Purchase Price": [avg_price],
                             "Number of Purchases": [total_purchases],
                             "Total Revenue": [total_revenue]}, columns= ["Number of Unique Items", "Average Purchase Price",
                            "Number of Purchases", "Total Revenue"])

#format
total_analysis_df.style.format({"Average Purchase Price": "${:.2f}", "Total Revenue": "${:.2f}"})
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#total number of players
full_count = purchase_data["SN"].nunique()

#total number of female players
female_count = purchase_data[purchase_data["Gender"] == "Female"]["SN"].nunique()

#total number of male players
male_count = purchase_data[purchase_data["Gender"] == "Male"]["SN"].nunique()

#total number of undisclosed or other
other_count = full_count - male_count - female_count

#calculate percentages
male_percent = ((male_count/full_count)*100)
female_percent = ((female_count/full_count)*100)
other_percent = ((other_count/full_count)*100)

#make a summary datafram
gender_summary_df = pd.DataFrame({"Gender": ["Male", "Female", "Other / Non-Disclosed"],
                                  "Total Count": [male_count, female_count, other_count],
                                   "Percentage of Players": [male_percent, female_percent, other_percent],
                                        }, columns = 
                                        ["Gender", "Total Count", "Percentage of Players"])
#set gender as Y index                                        
gender_summary_output_df = gender_summary_df.set_index("Gender")

#format the results
gender_summary_output_df.style.format({"Percentage of Players": "{:.2f}%"})
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
##purchase total by gender
gender_purchase_total = purchase_data.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Value")

#average purchase by gender
gender_average = purchase_data.groupby(["Gender"]).mean()["Price"].rename("Average Purchase Value")

#number of purchases by gender
gender_counts = purchase_data.groupby(["Gender"]).count()["Price"].rename("Purchase Count")


# Calculate average purchase per person
avg_total_pp = gender_purchase_total / gender_summary_output_df["Total Count"]

# make a summary dataframe
gender_data = pd.DataFrame({"Purchase Count": gender_counts, 
                            "Average Purchase Price": gender_average,
                            "Total Purchase Value": gender_purchase_total, 
                            "Avg Total Purchase per Person": avg_total_pp, })


#format
gender_data#.style.format({"Average Purchase Price": "{:.2f}"}, {"Average Total Purchase Per Person": "{:.2f}"})
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#assign age bins
age_bins = [0, 9.90, 14.90, 19.90, 24.9, 29.9, 34.90, 39.90, 9999999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", ">40"]

#Use Cut to categorize players
purchase_data["Age Range"]= pd.cut(purchase_data["Age"], age_bins, labels=group_names)
demographic_group =purchase_data.groupby("Age Range")
purchase_data.head()
```

```python
# Create age dataframe without duplicate entries
age_demographics = purchase_data[['Age','SN', 'Age Range']].drop_duplicates()


# calculate total count
age_count = age_demographics['Age Range'].value_counts()

# calculate ercentage of players
age_percentage = np.round(age_count / full_count * 100, decimals=2)

# Create and display summary data frame
age_demographics = pd.DataFrame({'Total Count': age_count, 'Percentage of Players': age_percentage})
age_demographics.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# categorize purchasing data using age bins
purchase_data['Age Group'] = pd.cut(purchase_data['Age'], age_bins, labels=group_names)

# purchase count
purchase_count = purchase_data.groupby(['Age Range']).count()['Purchase ID']

# average purchase price
avg_price = purchase_data.groupby(['Age Range']).mean()['Price']

# total purchase value
total_value = purchase_data.groupby(['Age Range']).sum()['Price']

# average total purchase per person
avg_per_player = total_value / age_demographics['Total Count']

# create and dispay summary data frame
age_purchase_df = pd.DataFrame({'Purchase Count': purchase_count.map('{:,}'.format),
                   'Average Purchase Price': avg_price.map('${:,.2f}'.format),
                   'Total Purchase Value': total_value.map('${:,.2f}'.format),
                   'Avg Total Purchase per Person': avg_per_player.map('${:,.2f}'.format)})
age_purchase_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#create variable to hold most items count
most_items = purchase_data.groupby(['Item ID', 'Item Name','Price'])['Price'].agg(['count','sum'])
most_items.columns = ['Purchase Count', 'Total Purchase Value']

# set the index back at zero 
most_items.reset_index(inplace=True)
most_items.set_index(['Item ID','Item Name'] ,inplace=True)

most_items = most_items[['Purchase Count', 'Price', 'Total Purchase Value']]

#Sort the purchase count column in descending order
most_popular_items = most_items.sort_values(by='Purchase Count', ascending=False)

#farmmat of the data base 
most_popular_items['Price'] = most_popular_items['Price'].map('${:,.2f}'.format)
most_popular_items['Total Purchase Value'] = most_popular_items['Total Purchase Value'].map('${:,.2f}'.format)

most_popular_items.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#sort in descending order 
most_profitable_items = most_items.sort_values(by='Total Purchase Value', ascending=False)

#create and display the summary dataframe
most_profitable_items['Price'] = most_profitable_items['Price'].map('${:,.2f}'.format)
most_profitable_items['Total Purchase Value'] = most_profitable_items['Total Purchase Value'].map('${:,.2f}'.format)
most_profitable_items.head()
```
